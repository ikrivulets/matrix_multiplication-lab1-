###  Импорт необходимых библиотек для парсинга и отрисовки графиков

```python
import re
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
matplotlib.rcParams['figure.figsize'] = (17.0, 10.0)
```

### Функция извлечения времени для соответствующего метода умножения из лога

```python
def extract_time(file_name, method_type):
    f = open(file_name,'r')
    found = re.findall(method_type + ': (.+?)\n', f.read())
    return [float(el) for el in found]
```

### Списки логов для анализа

```python
hundred_list = ['1000_1100_'+str(i)+'.txt' for i in xrange(1,6)]
thousand_list = ['1000_2000_'+str(i)+'.txt' for i in xrange(1,6)]
double_hundred_list = ['double_1000_1100_'+str(i)+'.txt' for i in xrange(1,6)]
double_thousand_list = ['double_1000_2000_'+str(i)+'.txt' for i in xrange(1,6)]
```

###  Сравнение метода умножения

```python
def compare_types(xlabel, type_list):    
    plt.ylabel('work_time(s)')
    plt.xlabel('matrix_size')
    finalSimple = [0]*101
    finalBlock = [0]*101
    finalSwapped = [0]*101
    cur_count = 0
    for file_name in type_list:
        timeSimple = extract_time(file_name, 'timeSimple')
        timeBlock = extract_time(file_name, 'timeBlock')
        timeSwapped = extract_time(file_name, 'timeSwapped')
        # count iterative average from logs
        for i in range(0,101):
            finalSimple[i] = (finalSimple[i] * cur_count + timeSimple[i]) / (cur_count + 1)
            finalBlock[i] = (finalBlock[i] * cur_count + timeBlock[i]) / (cur_count + 1)
            finalSwapped[i] = (finalSwapped[i] * cur_count + timeSwapped[i]) / (cur_count + 1)
        cur_count += 1
    
    plt.plot(xlabel,finalSimple, label='timeSimple')
    plt.plot(xlabel,finalBlock, label='timeBlock')
    plt.plot(xlabel,finalSwapped, label='timeSwapped')
        
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

### 1000-1100 для типа float:

```python
xlabel = range(1000,1101)
compare_types(xlabel, hundred_list)
```

### 1000-1100 для типа double:

```python
xlabel = range(1000,1101)
compare_types(xlabel, double_hundred_list)
```

### 1000-2000 для типа float:

```python
xlabel = range(1000,2001,10)
compare_types(xlabel,thousand_list)
```

### 1000-2000 для типа double:

```python
xlabel = range(1000,2001,10)
compare_types(xlabel,double_thousand_list)
```

### Анализ графиков:

Для timeSimple наблюдаем ярко выраженный выброс при размере 1024. Можно обьяснить тем, что много элементов попадает в один и тот же набор ячеек кэша, из-за чего получаем большое количество кэш-миссов. Также для него имеем много мелких скачков.

Алгоритмы Block и Swappped почти не подвержены данной проблеме, достаточно устойчивы и масштабируемы, разве что Blocked работает в приблизетельно полтора раза дольше на любом из графиков.

И для графиков с типом double получаем более сглаженные результаты в отличие от float.
